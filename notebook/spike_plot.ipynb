```python
import snntorch.spikeplot as splt
import matplotlib.pyplot as plt
import torch
import IPython.display as display
#  spike_data contains 128 samples, each of 100 time steps in duration
spike_data = torch.Tensor(100, 128, 1, 28, 28).uniform_() > 0.9
print(spike_data.size())

#  Index into a single sample from a minibatch
spike_data_sample = spike_data[:, 0, 0]
print(spike_data_sample.size())

#  Plot
fig, ax = plt.subplots()
anim = splt.animator(spike_data_sample, fig, ax)
#  Save as a gif
anim.save("spike_mnist.gif")
```

```python
import torch
import matplotlib.pyplot as plt

def gen_gaussian_distribution(len, mean, std):

    #extend the binning in oder to create the full gaussian and then cut the final part that goes to zeros
    bin = round(len*1.40)
    mean = round(mean*1.40)
    offset = round((bin - len)/2)
    x = torch.linspace(0, bin, bin)
    y = torch.exp(-((x - mean) ** 2) / (2 * std ** 2)) #multiply for the actual treshold max 0.5 is placeolder
    yy = y[offset:-offset]
    xx = x[offset:-offset]
    #show function, it will be removed when exported iside the network
    plt.plot(x, y)
    plt.ylim(0,1)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f"Plot of x and y {x.shape}")
    plt.show() 
    print(y)
    print(yy)
    plt.ylim(0,1)
    plt.plot(xx, yy)
    plt.xlabel('xx')
    plt.ylabel('yy')
    plt.title(f"Plot of x and y {xx.shape}")
    plt.show()
    #################
    return y

len = 20
m = len/2
var = 2
gen_gaussian_distribution(len, m, var)
```
